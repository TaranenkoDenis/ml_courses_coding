Decision Tree Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset['Salary']
```

Training the Decision Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
regressor = DecisionTreeRegressor(random_state=0) 
regressor.fit(X, y)

plt.figure(figsize=(12,8))
plot_tree(regressor, filled=True, feature_names=['Position'], rounded=True)
plt.show()
```

Predicting a new value

```python
y_pred = regressor.predict([[6.5]]) 
print("Predicted price: {}".format(y_pred))
```

Visualizing  the Decision Tree Regression results (high resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color='red', label='Actual Salaries')
plt.plot(X_grid, regressor.predict(X_grid), color='blue', label='Predicted Salaries')
plt.legend()
plt.title('Salary vs Position (Truth or Bluff)')
plt.show()
```
